```python
import pandas as pd
df = pd.read_csv('diabetes_data_upload.csv');
df.head()
```

```python
df.info()
```

```python
y = pd.factorize(df['class'])[0]

x = df.drop(['class'], axis=1)
x['Gender'] = pd.factorize(x['Gender'])[0]
x['Polyuria'] = pd.factorize(x['Polyuria'])[0]
x['Polydipsia'] = pd.factorize(x['Polydipsia'])[0]
x['sudden weight loss'] = pd.factorize(x['sudden weight loss'])[0]
x['weakness'] = pd.factorize(x['weakness'])[0]
x['Polyphagia'] = pd.factorize(x['Polyphagia'])[0]
x['Genital thrush'] = pd.factorize(x['Genital thrush'])[0]
x['visual blurring'] = pd.factorize(x['visual blurring'])[0]
x['Itching'] = pd.factorize(x['Itching'])[0]
x['Irritability'] = pd.factorize(x['Irritability'])[0]
x['delayed healing'] = pd.factorize(x['delayed healing'])[0]
x['partial paresis'] = pd.factorize(x['partial paresis'])[0]
x['muscle stiffness'] = pd.factorize(x['muscle stiffness'])[0]
x['Alopecia'] = pd.factorize(x['Alopecia'])[0]
x['Obesity'] = pd.factorize(x['Obesity'])[0]

x.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=21)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
dectree = DecisionTreeClassifier(max_depth=3, random_state=21)
dectree.fit(x_train,y_train)
dectree.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=((20,13)))
plot_tree(dectree, 
          filled=True,
          feature_names=x.columns)
plt.show()
```
